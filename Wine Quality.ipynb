```python
sns.countplot(x="quality", data=QT)
plt.title("Distribution of Wine Quality")
```

```python
sns.histplot(QT["alcohol"], bins=30)
plt.title("Distribution of Alcohol Content")
```

```python
sns.histplot(QT["density"], bins=20,kde=True)
plt.title("Distribution of Wine Density")
```

```python
sns.kdeplot(QT["alcohol"], fill=True)
plt.title("KDE Plot of Alcohol Content")
```

```python
sns.kdeplot(QT["quality"],fill=True)
plt.title("Distribution of pH  in Wine")
```

```python
sns.boxplot(x="quality", y="alcohol", data=QT)
plt.title("Alcohol Content by Wine Quality")
```

```python
sns.violinplot(x="quality", y="alcohol", data=QT)
plt.title("Violin Plot of Alcohol by Wine Quality")
```

```python
sns.scatterplot(x="alcohol", y="density", data=QT)
plt.title("Alcohol vs Density")
```

```python
sns.scatterplot(x="alcohol", y="quality", data=QT)
plt.title("Alcohol vs Wine Quality")
```

```python
sns.heatmap(QT.corr(), annot=False, cmap="coolwarm")
```

```python
sns.lineplot(x="quality", y="residual sugar", data=QT)
plt.title("Average Residual Sugar by Wine Quality")
```
