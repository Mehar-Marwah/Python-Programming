```python
#Question 7 
#Write a program that checks if a year is a leap year. A leap year is divisible by 4, but not by 100, unless it is also divisible by 400.
year=int(input("Enter the year"))
if year%4==0 or year%400==0 and year%100!=0:

    print("The year is a leap year")
else :
    print("The year is not a leap year")
    
```
