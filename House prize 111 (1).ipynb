```python
df_fence = df.dropna(subset=["Fence"])
sns.barplot( x="Fence", y="SalePrice",data=df_fence)
plt.title("Average House Price by Fence Type")
plt.show()
```

```python
sns.boxplot(x="Overall Qual",y="SalePrice",data=df)
plt.title("House Price Distribution by Overall Quality")
plt.show()
```

```python
sns.scatterplot(x="Gr Liv Area",y="SalePrice",data=df)
plt.title("Living Area vs Sale Price")
plt.show()
```

```python
sns.histplot(df["SalePrice"],bins=30,kde=True)
plt.title("Distribution of House Sale Prices")
plt.show()
```

```python
sns.violinplot(x="Garage Cars",y="SalePrice",data=df)
plt.title("Sale Price Distribution by Garage Capacity")
plt.show()
```

```python
sns.barplot(x="Overall Qual",y="SalePrice",data=df)
plt.title("Average Sale Price by Overall Quality")
plt.show()
```

```python
sns.countplot(x="Bedroom AbvGr", data=df)
plt.title("Count of Houses by Number of Bedrooms")
plt.show()
```

```python
sns.kdeplot(df["Gr Liv Area"], fill=True)
plt.title("Density Distribution of Living Area")
plt.show()
```

```python
sns.scatterplot(x="Gr Liv Area",y="SalePrice",hue="Fence",data=df)
plt.title("Living Area vs Sale Price by Fence")
plt.show()
```
