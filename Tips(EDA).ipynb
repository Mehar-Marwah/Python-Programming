```python
#Scatter PLot
df2 = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", hue="day", data=df)
plt.show()
```

```python
#Line Chart
df.sort_values("total_bill")
df.sort_values("tip")
sns.lineplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
#Relational PLot(relplot)
sns.relplot(x="total_bill", y="tip", hue="sex", col="smoker", data=df)
plt.show()
```

```python
#Histogram
sns.histplot(df["tip"], bins=50, kde=False)
plt.show()
```

```python
#kde
sns.kdeplot(df["tip"],fill=True)
plt.show()
```

```python
#Histogram
sns.histplot(df["tip"], bins=50, kde=True)
plt.show()
```

```python
#Bar Chart
sns.barplot(x="sex", y="tip", data=df)
plt.show()
```

```python
#Boxplot
sns.boxplot(x="smoker", y="total_bill", data=df)
```

```python
#Violin Plot
sns.violinplot(x="sex", y="tip", hue="smoker", data=df, split=False)
plt.show()
```

```python
#Heatmap    #Here Red colour represents- positive corerelation whereas blue colour shows negative corerelation 
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
#Pair plot
sns.pairplot(df, hue="size")
plt.show()
```

```python
#Countbar 
sns.countplot(x="size", hue="smoker", data=df)
plt.show()
```
