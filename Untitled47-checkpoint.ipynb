```python
# Looping through python
#if , elif , else statements 
#if = runs when if statement is true
#elif = runs when elif statement is true
#else = runs when both above statements are false 
#Q1 Looping through Python: Control flow statements: if, else, elif. Looping constructs: for and while loops with examples. Use of break, continue, and pass statements in loops
# Control Flow statements 
#example  
# To check and assign the score
score=int(input("Enter the score"))
if score>90:
    print("Your grade is A")
elif score>75:
    print("Your grade is B")
else:
    print("Your grade is C")
    
```

```python
#Looping constructs for and while loop
#For loop
for i in range(5):
    print("Number :",i)
```

```python
fruits=["apple","cherry","banana"]
for fruits in fruits:
    print(fruits)
```

```python
# While loop
count=0
while count<5:
    print("count:",count)
    count+=1
```

```python
# loop control statemts 
for i in range(10):
    if i==5:
        break 
    print(i)
```

```python
# continue 
for i in range (5):
    if i==2:
        continue
    print(i)
```

```python
#pass
for i in range(3):
    if i==1:
        pass #do nothing
    else:
        print("number:",i)
```
