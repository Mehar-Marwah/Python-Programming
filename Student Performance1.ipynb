```python
sns.scatterplot(x="reading score",y= "writing score",hue="gender",data=SP)
plt.title("Relationship Between Reading and Writing Scores by Gender")
plt.show()
```

```python
sns.relplot(x="reading score",y= "writing score",hue="gender",col="test preparation course",data=SP)
plt.show()
```

```python
SP = SP.sort_values(by="race/ethnicity")
sns.barplot(x="race/ethnicity",y="math score",hue="gender",data=SP)
plt.title("Average Math Score by Race/Ethnicity and Gender")
plt.show()
```

```python
sns.histplot(SP["writing score"],bins=25,kde=False)
plt.title("Distribution of Writing Scores")
plt.show()
```

```python
sns.kdeplot(SP["math score"],fill=True)
plt.title("Density Distribution of Math Scores")
plt.show()
```

```python
sns.histplot(SP["reading score"],bins=25,kde=False)
plt.title("Distribution of Reading Scores")
plt.show()
```

```python
SP = SP.sort_values(by="race/ethnicity")
SP["total_score"]=(SP["math score"] +SP["reading score"] +SP["writing score"])
```

```python
sns.barplot(x="race/ethnicity", y="total_score",data=SP)
plt.title("Average Total Score by Race/Ethnicity")
plt.show()
```

```python
sns.boxplot(x="gender", y="math score", data=SP)
plt.title("Math Score Distribution by Gender")
plt.show()
```
